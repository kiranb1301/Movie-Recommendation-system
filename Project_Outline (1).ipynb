## **Import Library**

```python
import pandas as pd 
import numpy as np
```

## **Import Data**

```python
df = pd.read_csv('https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Movies%20Recommendation.csv')
df.head()
```

## **Describe Data**

```python
df.info()
```

## **Data Visualization**

```python
df.describe()
```

## **Data Preprocessing**

```python
df.shape

df.columns

df_features = df[['Movie_Genre','Movie_Keywords','Movie_Tagline','Movie_Cast','Movie_Director']].fillna('')

df_features.shape

df_features
```

## **Define Target Variable (y) and Feature Variables (X)**

```python
x = df_features['Movie_Genre']+ '' + df_features['Movie_Keywords'] + ' ' + df_features['Movie_Tagline'] + ' ' + df_features['Movie_Cast'] + ' ' + df_features['Movie_Director']
x.shape
```

## **Train Test Split**

## **Modeling**

```python
from sklearn.feature_extraction.text import TfidfVectorizer 

Tfidf = TfidfVectorizer()

x = Tfidf.fit_transform(x)

x.shape
```

## **Model Evaluation**

```python
print(x)
```

```python
from sklearn.metrics.pairwise import  cosine_similarity

similarity_score = cosine_similarity(x)

similarity_score
```

```python
similarity_score.shape
```

## **Prediction**

```python
favourite_movie_name = input("enter your favorite name : ")

All_movies_title_list = df['Movie_Title'].tolist()

import difflib

movie_recomondation = difflib.get_close_matches(favourite_movie_name,All_movies_title_list)
print(movie_recomondation)



```

```python
close_match = movie_recomondation[0]
print(close_match)
```

```python
Index_of_close_match_movies = df[df.Movie_Title == close_match]['Movie_ID'].values[0]
print(Index_of_close_match_movies)
```

```python
Recomondation_score = list(enumerate(similarity_score[Index_of_close_match_movies]))
print(Recomondation_score)
```

```python
sorted_similar_movies = sorted(Recomondation_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies) 
```

```python
# top 30 movies recomonded movies
print("Top 30 movies suggested for you : \n ")
i = 1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = df[df.index == index]['Movie_Title'].values[0]
    if(i<31):
        print(i,'.',title_from_index)
        i += 1
```

```python
# Top 10 Movies recomonded for you 

Movie_Name = input("enter the movie name : ")

list_of_all_titles = df['Movie_Title'].tolist()

find_close_match = difflib.get_close_matches(Movie_Name, list_of_all_titles)

close_Match = find_close_match[1]

Index_of_Movie  = df[df.Movie_Title == close_Match]['Movie_ID'].values[0]

Recomondation_Score = list(enumerate(similarity_score[Index_of_Movie]))

sorted_similar_movies = sorted(Recomondation_Score, key= lambda x:x[1], reverse = True)

print("top 10 Movies suggested fo you : \n ")

i =1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = df[df.Movie_ID == index]['Movie_Title'].values
    
    if(i<11):
            print(i,'.',title_from_index)
            i += 1
    
```

## **Explaination**

A movie recommendation system project involves designing and implementing a system that suggests movies to users based on their preferences, historical behavior, and other relevant factors. Here's an overview of the key components and steps involved in such a project:

1. Data Collection:
Dataset Selection: Choose a suitable dataset containing information about movies, such as titles, genres, ratings, user reviews, cast, crew, release dates, etc.
Data Preprocessing: Clean the data, handle missing values, format data appropriately, and perform any necessary transformations to prepare it for analysis.


2. Exploratory Data Analysis (EDA):
Analyze the Data: Explore the dataset to gain insights into movie distributions, user preferences, rating distributions, correlations between features, etc.
Visualizations: Create visual representations (graphs, charts) to better understand the data.


3. Building the Recommendation System:
User Interface: Develop a user interface (web app, mobile app, or command-line interface) for users to interact with the recommendation system.
Integration with the Model: Implement the recommendation algorithms into the system to provide personalized movie suggestions based on user input or preferences.


6. Evaluation and Testing:
Evaluate Model Performance: Assess the effectiveness of the recommendation system using evaluation metrics like accuracy, precision, recall, RMSE (Root Mean Squared Error), etc.
Testing: Test the recommendation system with new or unseen data to ensure its robustness and generalizability.

A successful movie recommendation system not only provides accurate movie suggestions but also enhances user experience, engagement, and satisfaction by delivering personalized recommendations that align with users' tastes and preferences.
